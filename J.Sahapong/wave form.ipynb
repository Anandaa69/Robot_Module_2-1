```python
#Step 1: Basic Signal Generation Code

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def generate_signal(signal_type='sine', amplitude=1.0, frequency=1.0, period=1.0, sample_rate=1000):
    t = np.linspace(0, period, int(sample_rate * period), endpoint=False)

    if signal_type == 'sine':
        y = amplitude * np.sin(2 * np.pi * frequency * t)
    elif signal_type == 'square':
        y = amplitude * signal.square(2 * np.pi * frequency * t)
    elif signal_type == 'sawtooth':
        y = amplitude * signal.sawtooth(2 * np.pi * frequency * t)
    elif signal_type == 'triangle':
        y = amplitude * signal.sawtooth(2 * np.pi * frequency * t, 0.5)
    else:
        raise ValueError("Unsupported signal type.")

    return t, y

# Example visualization
t, y = generate_signal('square', amplitude=2.0, frequency=5.0, period=1.0)
plt.plot(t, y)
plt.title("Square Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
