```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure(figsize=(30, 7))

# กราฟที่ 1: bottle
df1 = pd.read_csv('data/tof_sensor_bottle_2.csv')
time1 = np.linspace(0, 30, len(df1))  # แกนเวลา 0-30 วินาที
plt.plot(time1, df1['tof'], label='Bottle', linestyle='-')

# กราฟที่ 2: ipad (flash angle)
df2 = pd.read_csv('data/tof_sensor_flash_angle_2.csv')
time2 = np.linspace(0, 30, len(df2))
plt.plot(time2, df2['tof'], label='iPad', linestyle='-')

# กราฟที่ 3: paper
df3 = pd.read_csv('data/tof_paper_2.csv')
time3 = np.linspace(0, 30, len(df3))
plt.plot(time3, df3['tof'], label='Paper', linestyle='-')

# ตั้งชื่อแกนและหัวเรื่อง
plt.xlabel('Time (seconds)')
plt.ylabel('TOF(mm)')
plt.title('3 subjects TOF Sensor Data Comparison')
plt.legend()
plt.tight_layout()
plt.show()

```
