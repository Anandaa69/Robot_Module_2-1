## lab6 Visual Feedback PID controller

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/faiijaran/Robot/Robot_Module_2-1/J.Fern/Lab6/save data/pid_data_from_robot_2.csv")
time = df["time"]

plt.figure(figsize=(10,6))
plt.plot(time, df["err_total"], label="err_total")
plt.plot(time, df["err_x"], label="err_x")
plt.plot(time, df["err_y"], label="err_y")
plt.xlabel("Time (s)")
plt.ylabel("Error")
plt.title("Error over Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(time, df["accumulate_err_x"], label="accumulate_err_x")
plt.plot(time, df["accumulate_err_y"], label="accumulate_err_y")
plt.plot(time, df["accumulate_err_total"], label="accumulate_err_total")
plt.xlabel("Time (s)")
plt.ylabel("Accumulated Error")
plt.title("Accumulated Errors over Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(time, df["controller_output_x"], label="controller_output_x")
plt.plot(time, df["controller_output_y"], label="controller_output_y")
plt.xlabel("Time (s)")
plt.ylabel("Controller Output")
plt.title("Controller Output over Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(time, df["gimbal_pitch_angle"], label="gimbal_pitch_angle")
plt.plot(time, df["gimbal_yaw_angle"], label="gimbal_yaw_angle")
plt.xlabel("Time (s)")
plt.ylabel("Angle (degrees)")
plt.title("Gimbal Angles over Time")
plt.legend()
plt.grid(True)
plt.show()
```
